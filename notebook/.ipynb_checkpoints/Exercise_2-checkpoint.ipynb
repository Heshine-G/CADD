### Task 4: Similarity Search

```python
row_names = ['Gefitinib', 'Erlotinib', 'Lapatinib', 'Afatinib', 'Osimertinib']
col_names = ['Gefitinib', 'Erlotinib', 'Lapatinib', 'Afatinib', 'Osimertinib']
Similarity_table = pd.DataFrame(index=row_names, columns=col_names)
```

```python
Similarity_table
```

#### Similarity search using Circular fingerprint - Morgan fingerprint and dice similarity metric

```python
for i in range(len(row_names)):
    reference_smile = smiles[i]
    reference_name = names[i]
    
    # Define reference molecule
    reference_molecule = Chem.MolFromSmiles(reference_smile)
    reference_morgan_fp = AllChem.GetMorganFingerprint(reference_molecule, radius=2)
    
    for j in range(len(col_names)):
        compound_name = names[j]
        
        # Skip comparison if row and column names match
        if i == j:
            Similarity_table.loc[row_names[i], col_names[j]] = '-'
            continue
        
        # Calculate similarity
        molecule = Chem.MolFromSmiles(smiles[j])
        morgan_fp = AllChem.GetMorganFingerprint(molecule, radius=2)
        dice_similarity = DataStructs.DiceSimilarity(reference_morgan_fp, morgan_fp)
        
        # Store similarity in the DataFrame
        Similarity_table.loc[row_names[i], col_names[j]] = dice_similarity
```

```python
Similarity_table
```

#### Similarity search using MACCS fingerprint and dice similarity metric

```python
for i in range(len(row_names)):
    reference_smile = smiles[i]
    reference_name = names[i]
    
    # Define reference molecule
    reference_molecule = Chem.MolFromSmiles(reference_smile)
    reference_maccs_fp = MACCSkeys.GenMACCSKeys(reference_molecule)
    
    for j in range(len(col_names)):
        compound_name = names[j]
        
        # Skip comparison if row and column names match
        if i == j:
            Similarity_table.loc[row_names[i], col_names[j]] = '-'
            continue
        
        # Calculate similarity
        molecule = Chem.MolFromSmiles(smiles[j])
        maccs_fp = MACCSkeys.GenMACCSKeys(molecule)
        similarity = DataStructs.DiceSimilarity(reference_maccs_fp, maccs_fp)
        
        # Store similarity in the DataFrame
        Similarity_table.loc[row_names[i], col_names[j]] = similarity
```

```python
Similarity_table
```

#### MACCS fingerprints capture predefined structural features that are often relevant to the studied property or activity, potentially resulting in higher similarity scores
#### when those features are shared among molecules. In contrast, Morgan fingerprints focus on local chemical environments within a specified radius,
#### which may not always capture the key structural elements driving similarity. Consequently, if the relevant features are not well-captured by the circular substructures
#### within the chosen radius, the similarity scores based on Morgan fingerprints may not accurately reflect the overall similarity between molecules.
#### In addition, the molecules Gefitinib and Afatinib have highest similarity compareed to other compounds.

```python
for mol in molecules:
    atom_indices = mol.GetSubstructMatch(substr_pattern)
    print(f"Matching atoms in molecule {Chem.MolToSmiles(mol)}:")
    for atom_idx in atom_indices:
        atom = mol.GetAtomWithIdx(atom_idx)
        print(f"Atom index: {atom.GetIdx()}, Atom symbol: {atom.GetSymbol()}")
        print()
```

### Task 8: Descriptor calculation for all the compounds

```python
molecular_weights = []
hb_acceptors = []
hb_donors = []
logp_value = []
for name, smile in zip(names, smiles):
    mol = Chem.MolFromSmiles(smile)
    molecular_weights.append(Descriptors.MolWt(mol))
    hb_acceptors.append(Descriptors.NumHAcceptors(mol))
    hb_donors.append(Descriptors.NumHDonors(mol))
    logp_value.append(Descriptors.MolLogP(mol))
descriptors_df = pd.DataFrame({
    'Compound': names,
    'Mol Weight': molecular_weights,
    'HB acceptor': hb_acceptors,
    'HB donor': hb_donors,
    'logP': logp_value
})
```

```python
descriptors_df
```

### Task 9: Lipinski's rule of five

```python
def lipinski_rules(mol_wt, hb_acceptor, hb_donor, logp):
    rules_fulfilled = 0
    
    if mol_wt <= 500:
        rules_fulfilled += 1
    if hb_acceptor <= 10:
        rules_fulfilled += 1
    if hb_donor <= 5:
        rules_fulfilled += 1
    if logp <= 5:
        rules_fulfilled += 1
    
    return rules_fulfilled

for index, row in descriptors_df.iterrows():
    rules_fulfilled = lipinski_rules(row['Mol Weight'], row['HB acceptor'], row['HB donor'], row['logP'])
    print(f"{row['Compound']}: {rules_fulfilled} rules fulfilled")
```

#### Test molecule that violates the rules

```python
test_molecule = {'Mol Weight': 600, 'HB acceptor': 12, 'HB donor': 3, 'logP': 6}
rules_fulfilled = lipinski_rules(test_molecule['Mol Weight'], test_molecule['HB acceptor'], test_molecule['HB donor'], test_molecule['logP'])
print(f"Test molecule: {4 - rules_fulfilled} rules violated")
```

#### Average and std dev calculation

```python
average = descriptors_df.select_dtypes(include='number').mean()
std_dev = descriptors_df.select_dtypes(include='number').std()

average_df = pd.DataFrame({'Compound': 'Average', **average}, index=[5])
std_dev_df = pd.DataFrame({'Compound': 'Std dev', **std_dev}, index=[6])

descriptors_and_values = pd.concat([descriptors_df, average_df, std_dev_df])
```

### Task 10: Active and Inactive EGFR compounds

#### Read the actives and inactives csv table

```python
actives = pd.read_csv("../data/Chembl_EGFR_actives_ML.csv")
inactives = pd.read_csv("../data/Chembl_EGFR_inactives_ML.csv")
```

```python
smiles_active = actives.canonical_smiles.tolist()
smiles_inactive = inactives.canonical_smiles.tolist()
```

#### Calculate descriptors for actives and inactives

```python
molecular_weights = []
hb_acceptors = []
hb_donors = []
logp_value = []
for smile in smiles_active:
    mol = Chem.MolFromSmiles(smile)
    molecular_weights.append(Descriptors.MolWt(mol))
    hb_acceptors.append(Descriptors.NumHAcceptors(mol))
    hb_donors.append(Descriptors.NumHDonors(mol))
    logp_value.append(Descriptors.MolLogP(mol))
actives_df = pd.DataFrame({
    'Mol Weight': molecular_weights,
    'HB acceptor': hb_acceptors,
    'HB donor': hb_donors,
    'logP': logp_value
})
```

```python
molecular_weights = []
hb_acceptors = []
hb_donors = []
logp_value = []
for smile in smiles_inactive:
    mol = Chem.MolFromSmiles(smile)
    molecular_weights.append(Descriptors.MolWt(mol))
    hb_acceptors.append(Descriptors.NumHAcceptors(mol))
    hb_donors.append(Descriptors.NumHDonors(mol))
    logp_value.append(Descriptors.MolLogP(mol))
inactives_df = pd.DataFrame({
    'Mol Weight': molecular_weights,
    'HB acceptor': hb_acceptors,
    'HB donor': hb_donors,
    'logP': logp_value
})
```

#### Calculate average and std dev for actives and inactives

```python
average_actives = actives_df.select_dtypes(include='number').mean()
average_inactives = inactives_df.select_dtypes(include='number').mean()

average_actives_df = pd.DataFrame(average_actives, columns=['Actives']).T
average_inactives_df = pd.DataFrame(average_inactives, columns=['Inactives']).T

all_average_df = pd.concat([average_actives_df, average_inactives_df])
all_average_df.loc['FDA Approved'] = average
all_average_df
```

```python
std_dev_actives = actives_df.select_dtypes(include='number').std()
std_dev_inactives = inactives_df.select_dtypes(include='number').std()

std_dev_actives_df = pd.DataFrame(std_dev_actives, columns=['Actives']).T
std_dev_inactives_df = pd.DataFrame(std_dev_inactives, columns=['Inactives']).T

all_std_dev_df = pd.concat([std_dev_actives_df, std_dev_inactives_df])
all_std_dev_df.loc['FDA Approved'] = std_dev
all_std_dev_df
```
